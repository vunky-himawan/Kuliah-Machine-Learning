```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = 'data/wbc.csv'
df = pd.read_csv(data)

df.head(10)
```

# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
# Menghapus variabel yang tidak dibutuhkan
df.drop(labels=['Unnamed: 32', 'id'], axis=1, inplace=True) # Menghapus kolom unnamed: 32 dan id

df.head()
```

# 2. Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

df.head()
```

# 3. Melakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()

df = df.astype(float)
df[:] = std.fit_transform(df[:])

display(df)
```

# 4. Melakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```
