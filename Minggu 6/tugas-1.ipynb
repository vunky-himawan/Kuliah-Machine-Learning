```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('Minggu 6/data/mushrooms.csv')

df.head(-1)
```

```python
df.columns
```

```python
feature_columns = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']

for column in feature_columns:
    print("===========================================")
    print(f"{column} ==> Missing zeros : {len(df.loc[df[column] == '?'])}")
```

```python
from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values='?', strategy='most_frequent', copy=False)

df[feature_columns] = fill_values.fit_transform(df[feature_columns])
```

```python
feature_columns = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']

for column in feature_columns:
    print("===========================================")
    print(f"{column} ==> Missing zeros : {len(df.loc[df[column] == '?'])}")
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)

X_encoded.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
```

```python
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
y_pred_rf = rf.predict(X_test)
```

```python
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print(f'Accuracy Decision Tree: {accuracy_dt}')
print(f'Accuracy Random Forest: {accuracy_rf}')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_dt = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}

grid_search_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train, y_train)
```

```python
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10]
}
```

```python
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)
```

```python
best_dt = grid_search_dt.best_estimator_
best_rf = grid_search_rf.best_estimator_
best_accuracy_dt = grid_search_dt.best_score_
best_accuracy_rf = grid_search_rf.best_score_

print(f'Best Accuracy Decision Tree: {best_accuracy_dt}')
print(f'Best Parameters Decision Tree: {grid_search_dt.best_params_}')
print(f'Best Accuracy Random Forest: {best_accuracy_rf}')
print(f'Best Parameters Random Forest: {grid_search_rf.best_params_}')
```

```python
y_pred_best_dt = best_dt.predict(X_test)
y_pred_best_rf = best_rf.predict(X_test)

best_accuracy_dt_test = accuracy_score(y_test, y_pred_best_dt)
best_accuracy_rf_test = accuracy_score(y_test, y_pred_best_rf)

print(f'Best Accuracy Decision Tree on Test Set: {best_accuracy_dt_test}')
print(f'Best Accuracy Random Forest on Test Set: {best_accuracy_rf_test}')
```
